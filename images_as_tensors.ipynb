```
from fastbook import *
import skimage.io as io
```

```
imag = Image.fromarray(io.imread('https://raw.githubusercontent.com/andandandand/images-for-colab-notebooks/master/3d-GrowCut-brain.PNG'))
imag
```

```
im = image2tensor(imag)
im
```

```
im.shape
```

```
show_image(im);
```

```
_, axs = subplots(1, 3)
for brain, ax, color in zip(im, axs, ('Reds','Greens','Blues')):
  show_image(255-brain, ax=ax, cmap=color)
```
